```python
# PANEL_XRANGE = (26, 106)
PANEL_XRANGE = (45, 106)
# PANEL_YRANGE = (30, 70)
PANEL_YRANGE = (40, 70)
PANEL_ZHOME = 9

def gen_bb_grid(*, nx=10, ny=5, xrange=PANEL_XRANGE, yrange=PANEL_YRANGE):
    xmin, xmax = xrange
    ymin, ymax = yrange

    # coords = [
    #     (xmin, ymin, PANEL_ZHOME),
    #     (xmin, ymax, PANEL_ZHOME),
    #     (xmax, ymin, PANEL_ZHOME),
    #     (xmax, ymax, PANEL_ZHOME),
    # ]
    coords = []
    stepx = (xmax - xmin) // (nx - 1)
    stepy = (ymax - ymin) // (ny - 1)
    
    for x in range(xmin, xmax + 1, stepx):
        for y in range(ymin, ymax + 1, stepy):
            coords.append((x, y, PANEL_ZHOME/2))

    return pd.DataFrame(coords, columns=['x', 'y', 'z'])


df = pd.concat([
    # gen_bb_grid(nx=2, ny=2, xrange=(60, 80), yrange=(40, 60)),
    # gen_bb_grid(nx=2, ny=5, xrange=(60, 65), yrange=(50, 55)),
    # gen_bb_grid(nx=5, ny=2, xrange=(70, 75), yrange=(60, 65)),
    # gen_bb_grid(nx=4, ny=4, xrange=(50, 75), yrange=(45, 65)),
    gen_bb_grid(nx=6, ny=5, xrange=(25, 100), yrange=(30, 70)),
    # gen_bb_grid(nx=4, ny=4, xrange=(50, 60), yrange=(50, 60)),
    # gen_bb_grid(nx=4, ny=4, xrange=(40, 90), yrange=(35, 65)),
    gen_bb_grid(nx=4, ny=3, xrange=(25, 50), yrange=(65, 75)),
    # gen_bb_grid(nx=3, ny=3, xrange=(40, 65), yrange=(45, 60)),
    # gen_bb_grid(nx=3, ny=3, xrange=(50, 70), yrange=(40, 55)),
    # gen_bb_grid(nx=3, ny=3, xrange=(55, 70), yrange=(48, 60)),
    pd.DataFrame([
            (100, 52, PANEL_ZHOME),
            (65, 30, PANEL_ZHOME),
            (25, 70, PANEL_ZHOME),
            # (65, 35, PANEL_ZHOME),
            # (25, 50, PANEL_ZHOME),
            # (105, 65, PANEL_ZHOME),
    ], columns=['x', 'y', 'z']),
])

fig = px.scatter(df, x='x', y='y', color='z', size='z')
fig.update_layout(
    width = 800,
    height = 500,
    title = "fixed-ratio axes"
)
fig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1,
  )
fig.show()


```
